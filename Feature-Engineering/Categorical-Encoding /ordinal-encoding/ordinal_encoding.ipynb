```python
# applying train_test_split functions to educations and review
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:3], df.iloc[:,-1], test_size=0.2)
```

```python
X_train
```

```python

from sklearn.preprocessing import OrdinalEncoder
```

```python
X_train
```

```python
# add that much list inside list total categorial data transform
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
oe.fit_transform(X_train)
oe.fit_transform(X_train)
```

```python
X_train = oe.transform(X_train)
```
