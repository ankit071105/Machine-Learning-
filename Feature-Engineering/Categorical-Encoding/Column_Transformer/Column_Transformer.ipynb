```python

import numpy as np
import pandas as pd
```

```python

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

```python
X_train
```

# **Without Column Transformer**

```python
# adding simple imputer to fever col
# filling the missing value with means of the fever column
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[['fever']])

# also the test data
X_test_fever = si.fit_transform(X_test[['fever']])

X_train_fever.shape
```

```python
X_train_fever # priniting after applying the simple imputer technique
# filling the missing value with means of the fever column
```

```python
# Ordinalencoding -> cough
oe = OrdinalEncoder(categories=[['Mild','Strong']]) # giving the rank means mild > Strong
X_train_cough = oe.fit_transform(X_train[['cough']])

# also the test data
X_test_cough = oe.fit_transform(X_test[['cough']])

X_train_cough.shape
```

```python
X_train_cough # priniting after applying the  Ordinalencoding technique
# ranking the strong as 1 and mild as 0
```

```python
# OneHotEncoding -> gender,city
ohe = OneHotEncoder(drop='first',sparse_output=False)
X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])

# also the test data
X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])

X_train_gender_city.shape
```

```python
X_train_gender_city  # priniting after applying the  OneHotEncoding technique
# [gender, city ,city , city]
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values

# also the test data
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age.shape
```

```python
X_train_age # Printing Extracting Age
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_gender_city,X_train_cough),axis=1)
# also the test data
X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough),axis=1)

X_train_transformed.shape
```

```python
X_train_transformed
#[Age , Fever , Gender , city1 , city 2 ,city 3 , cough]
```

# **# **With Column Transformer****

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    # (transformer name , transformer operator , list of col name )
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
# passthorugh - will not touch the col
# drop - will remove the col
```

```python
transformer.fit_transform(X_train)
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test)
```
