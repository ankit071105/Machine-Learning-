```python
dfs = []
```

```python
# correct code to fetch data through api
for i in range(1, 429):  # Adjust page range as needed
    url = f'https://api.themoviedb.org/3/movie/top_rated?api_key=c8e5ac89a9733f7dbceee867661323be&language=en-US&page={i}'
    response = requests.get(url)
    data = response.json().get('results', [])

    # Check if data is not empty
    if data:
      # fetching the data which are important
        temp_df = pd.DataFrame(data)[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        dfs.append(temp_df)
    else:
        print(f"No data found on page {i}")
```

```python
df = pd.concat(dfs, ignore_index=True)
```
