# **EDA – Exploratory Data Analysis**

1 Focus on Single Column
➔ Looking at one feature alone (e.g., Age, Gender).

2 Check Data Type
➔ Is it numerical (age, fare) or categorical (sex, class)?

3 Summary Statistics
➔ For numbers: mean, median, min, max, standard deviation
➔ For categories: count of each category

4 Visualisation
➔ For numerical: histogram, box plot
➔ For categorical: bar chart, count plot

5 Find Distribution Pattern
➔ Is it normal, skewed, or has outliers?

6 Detect Problems
➔ Missing values, wrong data types, unusual entries

7 No Relation with Other Variables
➔ Univariate means we don’t compare with other columns here.

# **Bivariate**

## **1. Scatterplot (Numerical - Numerical)**

```python
#** Bivariate **
```

```python
# Plot b/w total_bill and tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
# graph show as the bill increases the tip get increases
```

# **Multi-Variate**

```python
# same graph only identify b/w male and female by 'hue' functions

sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
# same graph only identify b/w male and female by 'hue' functions ,
# identify smoker by style functions by ( * and 0 )
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker')
```

```python
# same graph only identify b/w male and female by 'hue' functions ,
# identify smoker by style functions by ( * and 0 )
# size  for identifying its lunch ( big 0 ) dinner ( small 0 )
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='time')
```

## **2. Bar Plot (Numerical - Categorical)**

```python
titanic.head()
```

# **Bivariate**

```python
# Plot b/w age and Pclass (X-axis : Categorical ,  Y-axis : Numerical)
sns.barplot(x='Pclass', y='Age', data=titanic)
```

# **Multi-Variate**

```python
# Pclass & Age graph only identify b/w male and female by 'hue'  functions
sns.barplot(x='Pclass', y='Age', data=titanic, hue='Sex')
```

## **3. Box Plot (Numerical - Categorical)**

# **Bivariate**

```python
# Boxplot b/w sex and age
sns.boxplot(x='Sex', y='Age', data=titanic)
```

# **Multi-Variate**

```python
# Boxplot b/w sex , age and with hue functions Survived
sns.boxplot(x='Sex', y='Age', data=titanic, hue='Survived')
```

## **4. Distplot (Numerical - Categorical)**

```python
# kdeplot b/w age and survived (using pandas filter functions)
sns.kdeplot(data=titanic[titanic['Survived']==0], x='Age', label='Not Survived', color='black')
sns.kdeplot(data=titanic[titanic['Survived']==1], x='Age', label='Survived', color='blue')
```

```python
# histplot
sns.histplot(data=titanic[titanic['Survived']==0], x='Age', kde=True, label='Not Survived', color='red', alpha=0.3)
sns.histplot(data=titanic[titanic['Survived']==1], x='Age', kde=True, label='Survived', color='lightgreen', alpha=0.9)
```

## **5. HeatMap (Categorical - Categorical)**

```python
titanic.head(3)
```

```python
# pclass and Survived both Categorical
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
# light one shows the Survived people
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# percentage of people survived in all three class using pie chart.
(titanic.groupby('Sex')['Survived'].mean() * 100).plot(kind='bar')
```

```python
# percentage of people survived in all three class using pie chart.
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='pie',autopct='%.2f')
```

```python
# percentage of people survived in all places class using pie chart.
(titanic.groupby('Embarked')['Survived'].mean() * 100).plot(kind='pie',autopct='%.2f')
```
