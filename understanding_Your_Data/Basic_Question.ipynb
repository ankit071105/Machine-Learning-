```python
import pandas as pd
```

```python

df = pd.read_csv('train.csv')
```

```python
df['Survived'] = df['Survived'].fillna(0).astype(float)
```


# **1. How big is the data?**

```python
# checking the size of data
df.shape
```


# **2. How does the data look like?**

```python
# picking randomly value so that we can get everwhere data is same or different
df.sample(5)
```


# **3. What is the data type of cols?**

```python
# finding data type
df.info()
```


# **4. Are there any missing values?**

```python
# giving you total missing values of all the variables
df.isnull().sum()
```

# **5. How does the data look mathematically?**

```python
# defining the data in mathematical term
df.describe()
```

# **6. Are there duplicate values?**

```python
# finding duplicate value
df.duplicated().sum()
```
