```python
import numpy as np # importing numpy
import pandas as pd #importing pandas
```

```python
d = pd.read_csv('/content/placement.csv') #reading csv file from read_csv functions of pandas
```

```python
d.head() # give you the output how much data do you have with all col and rows
```

```python
d.info() # gives you information for all the rows & col and check whether valus is missed or not
```

```python
d.shape
```

```python
d=d.iloc[:,1:] # this code means i want all rows but removed col 1 (Unamed one)
```

```python
d.head()# give you the output how much data do you have with all col and rows
```

```python
import matplotlib.pyplot as plt # for graph  importing matplotlib libaray
```

```python
plt.scatter(d['cgpa'], d['iq']) # graph plotted between  CGPA and IQ
plt.xlabel('CGPA') # labeling the graph  x axis with CGPA
plt.ylabel('IQ') # labeling the graph  x axis with IQ
plt.title('CGPA vs IQ')  # Titling the graph with heading
plt.show()
```

```python
plt.scatter(d['cgpa'],d['iq'],c=d['placement']) # with cgpa & iq who got placement
```

```python
X = d.iloc[:,0:2]
y = d.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

xt,Xt,yt,Yt = train_test_split(X,y,test_size=0.1)
```

```python
xt
```

```python
yt
```

```python
Xt
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s = StandardScaler()
```

```python
xt = s.fit_transform(xt) # understand what values are there
```

```python
xt
```

```python
Xt = s.transform(Xt) # use to transform
```

```python
Xt
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
# model training training data
clf.fit(xt,yt)
```

```python
clf.predict(Xt)
```

```python
y_pred = clf.predict(Xt)
```

```python
Yt
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Yt,y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(xt, yt.values, clf=clf, legend=2)
```
